```python
books.rename(columns={
    "Book-Title": "title",
    "Book-Author": "author",
    "Year-Of-Publication": "year",
    "Publisher": 'publiser',
    "Image-URL-L": "image-url"
    
}, inplace = True)
```

```python
ratings = pd.read_csv('data/BX-Book-Ratings.csv', sep=";", on_bad_lines='skip', encoding='latin-1')
```

```python
ratings.head(2)
```

```python
ratings.rename(columns={
    "User-ID": "user_id",
    "Book-Rating": "rating"
}, inplace=True)
```

```python
ratings.head(2)
```

```python
ratings['user_id'].value_counts()
```

```python
ratings['user_id'].unique().shape
```

```python
x = ratings['user_id'].value_counts() > 200
```

```python
x[x].shape
```

```python
y = x[x].index
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
books.head(2)
```

```python
ratings_with_book = ratings.merge(books, on="ISBN")
```

```python
ratings_with_book.head()
```

```python
ratings_with_book.shape
```

```python
num_rating = ratings_with_book.groupby('title')['rating'].count().reset_index()
```

```python
num_rating.head()
```

```python
num_rating.rename(columns={'rating': 'num_of_rating'}, inplace=True)
```

```python
num_rating.head()
```

```python
final_ratings = ratings_with_book.merge(num_rating, on= 'title')
```

```python
final_ratings.head(2)
```

```python
final_ratings.shape
```

```python
final_ratings = final_ratings[final_ratings['num_of_rating']>=50]
```

```python
final_ratings.head()
```

```python
final_ratings.sample(10)
```

```python
final_ratings.shape
```

```python
final_ratings.drop_duplicates(['user_id', 'title'], inplace=True)
```

```python
final_ratings.shape
```

```python
final_ratings
```

```python
book_pivot = final_ratings.pivot_table(columns='user_id', index='title', values = 'rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```

```python
book_pivot.index
```

```python
books_name = book_pivot.index
```

```python
import pickle
pickle.dump(model, open('artifacts/model.pkl', 'wb'))
pickle.dump(books_name, open('artifacts/books_name.pkl', 'wb'))
pickle.dump(final_ratings, open('artifacts/final_ratings.pkl', 'wb'))
pickle.dump(book_pivot, open('artifacts/book_pivot.pkl', 'wb'))
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6)
    
    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```

```python
book_name = 'A Bend in the Road'
recommend_book(book_name)
```
